## Single sequence, printing the result

```python
seq = "GGGCGGCUAGCUCAGCGGAAGAGCGCUCGCCUCACACGCGAGAGGUCGUAGGUUCAAGUCCUACGCCGCCCACCA"

SQUARNA.Predict(inputseq = seq)
```

## With Rfam search enabled

```python
#Requires (1) an independent Infernal installation and (2) running SQUARNA-build-rfam in the terminal
seq = "GGGCGGCUAGCUCAGCGGAAGAGCGCUCGCCUCACACGCGAGAGGUCGUAGGUUCAAGUCCUACGCCGCCCACCA"

SQUARNA.Predict(inputseq = seq, rfam = True)
```

## With G-quadruplex pattern recognition enabled

```python
seq = "GCGCCAUGGGCCAUUGGGUGGGAUCUGGGAUGGCGC"

SQUARNA.Predict(inputseq = seq, g4 = True)
```

## With structural restraints

```python
seq  = "GCGCCAUGGGCCAUUGGGUGGGAUCUGGGAUGGCGC"
rstr = "..........((...))..................."

SQUARNA.Predict(inputseq = seq, inputrestr = rstr)
```

## Single sequence, write the result to a file

```python
seq = "GGGCGGCUAGCUCAGCGGAAGAGCGCUCGCCUCACACGCGAGAGGUCGUAGGUUCAAGUCCUACGCCGCCCACCA"

outpfile = "demo.txt"

with open(outpfile,"w") as file:
    SQUARNA.Predict(inputseq = seq, write_to = file)
```

## Single sequence, capture the result into a string

```python
import io
seq = "GGGCGGCUAGCUCAGCGGAAGAGCGCUCGCCUCACACGCGAGAGGUCGUAGGUUCAAGUCCUACGCCGCCCACCA"

with io.StringIO() as stream:
    SQUARNA.Predict(inputseq = seq, write_to = stream)
    result = stream.getvalue()

lines = result.split('\n')
print(lines)
```
